# Preliminary testing of the C-band scatterometer

I made 4 1-minute long runs:

* Calm, smooth surface, no wind or paddle waves
* Wind ~5 m/s, no paddle waves
* No wind, monochromatic paddle waves with 1 s period and 0.05 m amplitude
* Wind ~5 m/s and paddle waves as above

Below I plot:

1. Hpol
2. Vpol
3. ref

variables from the HDF5 files, each plot showing the 1-minute average in each run.
The y-range is always the same in each figure to allow seeing the difference in magnitude between runs,
but the ranges vary between variables.

```python
import matplotlib.pyplot as plt
import numpy as np
import radar
plt.rcParams.update({'font.size': 16})
```

```python
path = '/home/milan/Work/sustain/c-band-scatterometer/data/test_20210311'

files = [
    'radar_0Hz.h5',
    'radar_10Hz.h5',
    'radar_0Hz_paddle_1s_5cm.h5',
    'radar_10Hz_paddle_1s_5cm.h5'
]

runs = []
for filename in files:
    runs.append(radar.load(path + '/' + filename))
```

## Hpol, real component

```python
fig = plt.figure(figsize=(16, 12))
ax1 = plt.subplot2grid((2, 2), (0, 0))
ax2 = plt.subplot2grid((2, 2), (0, 1))
ax3 = plt.subplot2grid((2, 2), (1, 0))
ax4 = plt.subplot2grid((2, 2), (1, 1))

ax1.set_title('No wind, no paddle')
ax2.set_title('Wind, no paddle')
ax3.set_title('No wind, paddle')
ax4.set_title('Wind, paddle')

ax1.plot(runs[0][2].real.mean(axis=0))
ax2.plot(runs[1][2].real.mean(axis=0))
ax3.plot(runs[2][2].real.mean(axis=0))
ax4.plot(runs[3][2].real.mean(axis=0))

for ax in [ax1, ax2, ax3, ax4]:
    ax.set_ylim(1.14e8, 1.21e8)
    ax.grid()

fig.suptitle('Hpol, real component, 1-minute average')
```

## Time

Time seems to have odd jumps in value. Is this expected?

Note that the time shown here is offset by the initial value (`time - time[0]`) so that the time is in seconds, starting from 0.

```python
fig = plt.figure(figsize=(16, 12))
ax1 = plt.subplot2grid((2, 2), (0, 0))
ax2 = plt.subplot2grid((2, 2), (0, 1))
ax3 = plt.subplot2grid((2, 2), (1, 0))
ax4 = plt.subplot2grid((2, 2), (1, 1))

ax1.set_title('No wind, no paddle')
ax2.set_title('Wind, no paddle')
ax3.set_title('No wind, paddle')
ax4.set_title('Wind, paddle')

ax1.plot(runs[0][0])
ax2.plot(runs[1][0])
ax3.plot(runs[2][0])
ax4.plot(runs[3][0])

for ax in [ax1, ax2, ax3, ax4]:
    ax.grid()

fig.suptitle('Time since start of the recording [s]')
```

## Time step

And in the case where time is continuous (no jumps), like in the calm or wind+paddle cases, the time step seems to vary. Is this expected?

```python
fig = plt.figure(figsize=(16, 6))
time = runs[0][0]
time_step = np.diff(time)
plt.plot(time[:-1], time_step)
plt.xlabel('Time [s]')
plt.ylabel('Time step [s]')
plt.grid()
fig.suptitle('Calm')
```

```python
fig = plt.figure(figsize=(16, 6))
time = runs[3][0]
time_step = np.diff(time)
plt.plot(time[:-1], time_step)
plt.xlabel('Time [s]')
plt.ylabel('Time step [s]')
plt.grid()
fig.suptitle('Wind + paddle')
```

## Takeaways

* Re(H):
  - Flat and low in calm conditions
  - Moderate and variable in wind-only and paddle-only conditions
  - Higher and variable in wind+padde conditions
* Im(H):
  - Low in no paddle, regardless of wind
  - High with paddle, regardless of wind
* Re(V):
  - Low in calm conditions
  - Moderate in paddle conditions, regardless of wind
  - High in wind-only conditions
* Im(V):
  - Low negative in all conditions except wind+paddle
  - High negative in wind+paddle conditions
* Re(ref)
  - High negative in all conditions except paddle-only
  - Low negative in paddle-only conditions
* Im(ref)
  - Opposite of Re(ref)
